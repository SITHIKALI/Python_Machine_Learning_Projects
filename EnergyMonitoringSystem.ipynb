```python
file_name = "EnergyMeter.csv"
names = ['voltage','current','power','class']
dataset = read_csv(file_name, skiprows=1, names=names)

# Convert the feature columns to numeric type
dataset['voltage'] = dataset['voltage'].astype(float)
dataset['current'] = dataset['current'].astype(float)
dataset['power'] = dataset['power'].astype(float)
```

```python
# Display the dataset with first 10 rows
display(dataset.head(10))
```

```python
print(type(dataset))
```

```python
# seggregation
x = dataset.values[:,0:3] # independent variable
y = dataset.values[:,3] # dependent variable
x_train,x_val,y_train,y_val = train_test_split(x,y,test_size=0.20,random_state = 1)
```

```python
model = SVC(gamma='auto')
model.fit(x_train,y_train)
result = model.score(x_val,y_val)
print(result)
```

```python
# Prediction
value = [[236.0,5.84,1378.24]]
predictions = model.predict(value)
print(predictions[0])
```
